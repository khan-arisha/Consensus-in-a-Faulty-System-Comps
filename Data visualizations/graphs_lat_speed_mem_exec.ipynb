```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get this data from csv: latency.csv and speed.csv
nodes = [3, 4, 5, 6, 7, 8, 9, 10]
hotstuff_speed = [0.150504, 0.160749, 0.174411, 0.221783, 0.259649, 0.36653, 0.372377, 0.381448]
raft_speed = [0.224677, 0.296173, 0.287402, 0.429355, 0.560944, 0.713363, 0.775183, 0.828331]
paxos_speed = [0.011739, 0.011979, 0.012612, 0.013169, 0.013423, 0.013727, 0.014194, 0.015659]


plt.figure(figsize=(8, 5))
plt.plot(nodes, hotstuff_speed, marker='o', label='Hotstuff', linestyle='-',  color='#f54242')
plt.plot(nodes, raft_speed, marker='s', label='RAFT', linestyle='--' , color='#42f566')
plt.plot(nodes, paxos_speed, marker='^', label='Leaderless Byzantine Paxos', linestyle=':', color = '#42b0f5')

# Labels and title
plt.xlabel('Number of Nodes')
plt.ylabel('Speed to Reach Consensus (s)')
plt.title('Consensus Speed Comparison')
plt.legend()


plt.savefig('/Users/arishakhan/Downloads/consensus_Speed.png', dpi=300)
plt.show()
```
