```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from google.colab import drive

# url = "https://docs.google.com/spreadsheets/d/1SlPjsDGDTheFi9RfCygOY2kgx5U54Uny0kMfLd13-Rw/gviz/tq?tqx=out:csv&gid=358622378"
url = "performance.csv"

df = pd.read_csv(url)
df
```

```python
algorithm = df["Executed MPI Program"]
time = df["Execution Time (ms)"]
memory = df["Memory Usage (KB)"]

plt.figure(figsize=(10, 6))

# Plot Execution Time
plt.plot(algorithm, time, marker='o', linestyle='-', label="Execution Time (ms)")

# Plot Memory Usage
plt.plot(algorithm, memory, marker='s', linestyle='--', label="Memory Usage (KB)")

plt.title('Execution Time and Memory Usage')
plt.xlabel('Algorithm')
plt.ylabel('Value')
plt.xticks(rotation=9, ha='center')
plt.legend()  # Add a legend to differentiate the lines
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
# Average Memory Usage per Algorithm
df_grouped_memory = df.groupby('Executed MPI Program')['Memory Usage (KB)'].mean()

plt.figure(figsize=(10, 6))
df_grouped_memory.plot(kind='line', marker='o', linestyle='-', color='violet')  # Line graph
plt.xlabel('Executed MPI Program')
plt.ylabel('Average Memory Usage (KB)')
plt.title('Average Memory Usage per Algorithm')
plt.xticks(rotation=9, ha='center')
plt.grid(True)
plt.show()

# Average Execution Time per Algorithm
df_grouped_time = df.groupby('Executed MPI Program')['Execution Time (ms)'].mean()

plt.figure(figsize=(10, 6))
df_grouped_time.plot(kind='line', marker='s', linestyle='-', color='blue')  # Line graph
plt.xlabel('Executed MPI Program')
plt.ylabel('Average Execution Time (ms)')
plt.title('Average Execution Time per Algorithm')
plt.xticks(rotation=9, ha='center')
plt.grid(True)
plt.show()
```
